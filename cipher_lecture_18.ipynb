```python
#Creating a simple line plot with matplotlib

import matplotlib.pyplot as plt
#Data
x=[1,2,3,4,5]
y=[2,3,5,7,11]

#Creating a line plot
plt.plot(x,y,linestyle="dotted")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple line plot')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple line plot')
plt.show()
```

```python
plt.plot(x,y,linestyle="dashed")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple line plot')
plt.show()
```

```python
plt.plot(x,y,ls=":") #Shortcut   (use '--' for dashed, '-' for solid, '-.' for dashdot)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple line plot')
plt.show()
```

```python
plt.plot(x,y,linestyle="dashed",color="r",linewidth="10") #can also use hexcodes to provide colors
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple line plot')
plt.show()
```

```python
import numpy as np
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
plt.plot(y1,marker="o",mec="g",ms=20) #ms gives marker size, mec is for marker color(marker outside interface)
plt.plot(y2,marker="s",mfc="y") #mfc changes solid marker color
plt.show()
```

Creating a simple scatter plot with Matplotlib

```python
#Data

x=[1,2,3,4,5]
y=[2,3,5,7,11]

#Creating scatter plot
plt.scatter(x,y,color="r")
plt.xlabel('Sales')
plt.ylabel('Prices')
plt.title('Simple Scatter Plot')
plt.show()
```

Creating a simple bar plot with matplotlib

```python
#Data

categories=['A','B','C','D']
values=[4,7,1,8]

#creating a bar plot
plt.bar(categories,values,color="y",width=0.1) #barh for horizontal
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot')
plt.show()
```

Creating simple histogram with matplotlib

```python
data=[1,2,2,3,3,3,4,4,4,4]

#creating a histogram
plt.hist(data,bins=4)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram')
plt.show()
```

```python
#Creating subplots with matplotlib
x=[1,2,3,4,5]
y1=[2,3,5,7,11]
y2=[1,4,6,8,10]

#creating subplots
fig,axs=plt.subplots(2)
axs[0].plot(x,y1)
axs[0].set_title('First Plot')
axs[1].plot(x,y2,'tab:orange')
axs[1].set_title('Second Plot')

#Displaying the plot
plt.tight_layout()
plt.show()
```

Adding annotations with matplotlib

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]
#Creating a plot with annotations
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with annotations')
plt.annotate('Peak',xy=(5,11),xytext=(4,10),arrowprops=dict(facecolor='yellow',shrink=0.09))
plt.show()
```

Creating a simple line plot with seaborn

```python
import seaborn as sns #Always built on top of matplotlib, used to beautify
import matplotlib.pyplot as plt


sns.lineplot(x=x,y=y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple line plot seaborn')
plt.show()
```

```python
#Creating simple scatterplot with seaborn
sns.scatterplot(x=x,y=y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple scatter plot seaborn')
plt.show()
```

Creating a simple bar plot with seaborn

```python
#Data
categories=['a','b','c','d']
values=[4,7,1,8]

#Creating a bar plot
sns.barplot(x=categories,y=values)
plt.xlabel('Categories')
plt.ylabel('Value')
plt.title('Simple Bar Plot with Seaborn')
plt.show()
```

Create a simple histogram with seaborn

```python
#Data
data=[1,2,2,3,3,3,4,4,4,4]

#Creating a histogram with seaborn
sns.histplot(data,bins=4)
plt.xlabel=('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram with seaborn')
plt.show()
```

Creating pairplot with seaborn


```python
from sklearn.datasets import load_iris

iris=load_iris()
iris_data=sns.load_dataset("iris")


#Creating a pairplot
sns.pairplot(iris_data, hue="species")
plt.title('Pairplot with seaborn')
plt.show()
```
